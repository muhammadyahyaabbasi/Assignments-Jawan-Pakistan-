## Write a python program to sun all the numeric items in a list

```python

my_list1 = [10, 20, 30, 40, 50]

# Calculate sum
total = sum(my_list1)

print("Sum of all numeric items:", total)
```

```python
#Using Loop
my_list2 = [10, 20, 30, 40, 50]

total = 0

for item in my_list2:
    total += item

print("Sum of all numeric items:", total)
```

```python
#Contains Non-Numeric Values

my_list3 = [10, "hello", 20, 30.5, "world", 40]

total = 0

for item in my_list3:
    if isinstance(item, (int, float)):
        total += item

print("Sum of numeric items:", total)
```

## Write a python program to get the largest from a numeric list

```python

numbers = [10, 45, 23, 89, 67]

# Find largest number
largest = max(numbers)

print("The largest number is:", largest)
```

```python
# using loop
numbers1 = [10, 45, 23, 89, 67]

largest = numbers1[0]

for num in numbers:
    if num > largest:
        largest = num

print("The largest number is:", largest)
```

```python
# user input
numbers = list(map(int, input("Enter numbers separated by space: ").split()))

largest = max(numbers)

print("The largest number is:", largest)
```

## Write a python program that print out all the elements of the list that are less than 5

```python

num1 = [1, 7, 3, 9, 4, 6, 2]

# Print elements less than 5
for num in num1:
    if num < 5:
        print(num)
```

```python
num2 = [1, 7, 3, 9, 4, 6, 2]

less_than_five = [num for num in num2 if num < 5]

print("Elements less than 5:", less_than_five)
```

```python
#user input 
num3 = list(map(int, input("Enter numbers separated by space: ").split()))

for num in num3:
    if num < 5:
        print(num)
```
