## 1. Calculator using Python

```python
print("Simple Calculator")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")
print("5. Power")

choice = int(input("Enter your choice (1-5): "))

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

if choice == 1:
    print("Result:", num1 + num2)

elif choice == 2:
    print("Result:", num1 - num2)

elif choice == 3:
    print("Result:", num1 * num2)

elif choice == 4:
    if num2 != 0:
        print("Result:", num1 / num2)
    else:
        print("Division by zero not allowed")

elif choice == 5:
    print("Result:", num1 ** num2)

else:
    print("Invalid choice")
```

## 2. Check if there is any numeric value in list (using for loop)

```python
my_list = ["Ali", "Ahmed", 25, "Sara", "Pakistan"]

found = False

for item in my_list:
    if isinstance(item, (int, float)):
        found = True
        break

if found:
    print("Numeric value exists in the list")
else:
    print("No numeric value found")
```

## 3. Add a key to a dictionary

```python
my_dict = {"name": "Ali", "age": 22}

# Adding new key
my_dict["city"] = "Karachi"

print(my_dict)
```

## 4. Sum all numeric items in a dictionary

```python
my_dict = {"a": 10, "b": 20, "c": 30, "d": "hello"}

total = 0

for value in my_dict.values():
    if isinstance(value, (int, float)):
        total += value

print("Sum of numeric values:", total)
```

## 5. Identify duplicate values from list

```python
my_list = [1, 2, 3, 4, 2, 5, 3, 6, 1]

duplicates = []

for item in my_list:
    if my_list.count(item) > 1 and item not in duplicates:
        duplicates.append(item)

print("Duplicate values:", duplicates)
```

## 6. Check if a key already exists in a dictionary

```python
my_dict = {"name": "Ali", "age": 22, "city": "Karachi"}

key = input("Enter key to check: ")

if key in my_dict:
    print("Key exists in dictionary")
else:
    print("Key does not exist")
```
