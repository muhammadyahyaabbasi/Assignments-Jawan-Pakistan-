```python
# 7. Marksheet with Grades

print("Enter marks of 5 subjects")

sub1 = float(input("Math: "))
sub2 = float(input("Science: "))
sub3 = float(input("PST: "))
sub4 = float(input("Urdu: "))
sub5 = float(input("Computer: "))

total = sub1 + sub2 + sub3 + sub4 + sub5
percentage = (total / 500) * 100

print("Total Marks:", total)
print("Percentage:", percentage)

if percentage >= 80:
    print("Grade: A+")
elif percentage >= 70:
    print("Grade: A")
elif percentage >= 60:
    print("Grade: B")
elif percentage >= 50:
    print("Grade: C")
else:
    print("Grade: Fail")
```
