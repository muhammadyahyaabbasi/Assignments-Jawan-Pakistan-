```python
import math


print("===== Made by Muhammad Yahya =====")

print("===== Advanced Calculator =====")

while True:
    print("\nSelect Operation:")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Power (x^y)")
    print("6. Square Root")
    print("7. Exit")

    choice = input("Enter choice (1-7): ")

    if choice == '7':
        print("Calculator Closed ")
        break

    # Operations requiring two numbers
    if choice in ('1', '2', '3', '4', '5'):
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))

        if choice == '1':
            print("Result =", num1 + num2)

        elif choice == '2':
            print("Result =", num1 - num2)

        elif choice == '3':
            print("Result =", num1 * num2)

        elif choice == '4':
            if num2 != 0:
                print("Result =", num1 / num2)
            else:
                print("Error Cannot divide by zero")

        elif choice == '5':
            print("Result =", num1 ** num2)

    # Square Root (one number)
    elif choice == '6':
        num = float(input("Enter number: "))
        if num >= 0:
            print("Result =", math.sqrt(num))
        else:
            print("Error Cannot find square root of negative number")

    else:
        print("Invalid Input")


        
```

```python

print("===== Made by Muhammad Yahya =====")
print("===== Student Marksheet =====")

name = input("Enter Student Name: ")

# 5 subjects ke marks input
sub1 = float(input("Enter marks of Subject Math: "))
sub2 = float(input("Enter marks of Subject Science: "))
sub3 = float(input("Enter marks of Subject PST: "))
sub4 = float(input("Enter marks of Subject Urdu: "))
sub5 = float(input("Enter marks of Subject Physics: "))

# Total aur Percentage
total = sub1 + sub2 + sub3 + sub4 + sub5
percentage = (total / 500) * 100

# Grade System
if percentage >= 90:
    grade = "A+"
elif percentage >= 80:
    grade = "A"
elif percentage >= 70:
    grade = "B"
elif percentage >= 60:
    grade = "C"
elif percentage >= 50:
    grade = "D"
else:
    grade = "Fail"

# Output
print("\n===== Marksheet =====")
print("Student Name:", name)
print("Total Marks:", total, "/ 500")
print("Percentage:", percentage, "%")
print("Grade:", grade)
```
